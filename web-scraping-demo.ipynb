```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'} 
webpage = requests.get('https://www.ambitionbox.com/list-of-companies?page=1', headers=headers).text
```

```python
soup = BeautifulSoup(webpage, 'lxml')
```

```python
company = soup.find_all('div', class_='company-content-wrapper')
```

```python
name = []
rating = []
reviews = []
ctype = []
hq = []
how_old = []
no_of_employees = []

for i in company:
    name.append(i.find('h2').text.strip())
    rating.append(i.find('p', class_='rating').text.strip())
    reviews.append(i.find('a', class_='review-count').text.strip())   
    ctype.append(i.find_all('p', class_='infoEntity')[0].text.strip())
    hq.append(i.find_all('p', class_='infoEntity')[1].text.strip())    
    how_old.append(i.find_all('p', class_='infoEntity')[2].text.strip())    
    no_of_employees.append(i.find_all('p', class_='infoEntity')[3].text.strip())
    
```

```python
f = {'name':name, 'rating':rating, 'reviews':reviews, 'type':ctype, 'hq':hq, 'old':how_old, 'employees':no_of_employees}
```

```python
df = pd.DataFrame(f)
```

```python
import numpy as np
```

```python
fdf = pd.DataFrame()

for j in range(1, 100):
    url = 'https://www.ambitionbox.com/list-of-companies?page={}'.format(j)
    
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'} 
    webpage = requests.get(url, headers=headers).text
    
    soup = BeautifulSoup(webpage, 'lxml')
    company = soup.find_all('div', class_='company-content-wrapper')
    name = []
    rating = []
    reviews = []
    ctype = []
    hq = []
    how_old = []
    no_of_employees = []

    for i in company:
        name.append(i.find('h2').text.strip())
        rating.append(i.find('p', class_='rating').text.strip())
        reviews.append(i.find('a', class_='review-count').text.strip())   
        ctype.append(i.find_all('p', class_='infoEntity')[0].text.strip())
        try:
            hq.append(i.find_all('p', class_='infoEntity')[1].text.strip())
        except:
            hq.append(np.nan)
        try:
            how_old.append(i.find_all('p', class_='infoEntity')[2].text.strip())
        except:
            how_old.append(np.nan)
        try:
            no_of_employees.append(i.find_all('p', class_='infoEntity')[3].text.strip())
        except:
            no_of_employees.append(np.nan)
    f = {'name':name, 'rating':rating, 'reviews':reviews, 'type':ctype, 'hq':hq, 'old':how_old, 'employees':no_of_employees}
    df = pd.DataFrame(f)
    fdf = fdf.append(df, ignore_index=True)
```
